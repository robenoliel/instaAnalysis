```python
import selenium.webdriver as webdriver                #Acesso a página web
from instaReader import instaReader
import mysql.connector
import pandas as pd
```

```python
def drop_tables():
    return """
drop tables if exists Posts, Users;
"""
def create_users_table():
    return """
create table Users (
    username varchar(30) primary key,
    followers int unsigned not null
);
"""
def create_posts_table():
    return """
create table Posts (
	insta_url varchar(11) not null unique primary key,
    date date not null,
	time time not null,
    likes int unsigned not null,
	popularity float not null,
    text varchar(2200),
    username varchar(30),
    
    constraint FK_username foreign key (username)
    references Users (username)
    on delete cascade
    on update cascade
);  
    """

def insert_str(tb, df):
    string = 'insert into {0} values '.format(tb)
    for i, row in df.iterrows():
        columns = tuple(row)
        string+= '{0},'.format(columns)
    string = string[:-1]+';'
    return string

user = 'root'
passw = 'RoBenoBagda_99'
host = 'localhost'
db = 'pelsias_social'
db_connection = mysql.connector.connect(
  	host=host,
  	user=user,
  	passwd=passw,
    database=db)
db_cursor = db_connection.cursor()

db_cursor.execute(drop_tables())
db_cursor.execute(create_users_table())
db_cursor.execute(create_posts_table())
```

```python
user = 'pelsias.ufrj'
limit = 100
driver = webdriver.Firefox(executable_path = 'C:\\Users\\roben\\geckodriver.exe')
insta = instaReader(driver, user,'pels2019ufrj')
user = 'pelsias.ufrj'
df = insta.scrapPosts(number_of_posts = limit, user = user)
df['time'] = df['date'].str[11:19]
df['date'] = df['date'].str[0:10]
df = df.reindex(columns= ['code', 'date', 'time','likes','popularity','text','user'])
df_users = insta.get_users(users = [user])
print(df_users)
df_users = df_users.reindex(columns= ['user','followers'])
db_cursor.execute(insert_str('Users',df_users))
df = df.dropna()
db_cursor.execute(insert_str('Posts',df))
db_connection.commit()
```

```python
db_connection.close()
driver.quit()
```
